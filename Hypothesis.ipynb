```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given values
population_mean = 82
sample_mean = 90
population_std_dev = 20
sample_size = 81
alpha = 0.05

# Step 1: Calculate the Z-statistic
z_stat = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Step 2: Critical value for one-tailed test at alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha)

# Step 3: Make the decision
if z_stat > critical_value:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

# Step 4: Plotting the normal distribution
x = np.linspace(population_mean - 4 * population_std_dev, population_mean + 4 * population_std_dev, 1000)
y = stats.norm.pdf(x, population_mean, population_std_dev / np.sqrt(sample_size))

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='blue')
plt.fill_between(x, 0, y, where=(x >= sample_mean), color='red', alpha=0.5, label=f'Rejection Region (z > {critical_value:.2f})')
plt.axvline(x=sample_mean, color='green', linestyle='dashed', label=f'Sample Mean = {sample_mean}')
plt.axvline(x=population_mean, color='orange', linestyle='dashed', label=f'Population Mean = {population_mean}')
plt.title('Z-Test: Hypothesis Testing')
plt.xlabel('Score')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Output the results
print(f"Calculated Z-Statistic: {z_stat:.2f}")
print(f"Critical Z-Value: {critical_value:.2f}")
print(f"Decision: {decision}")
```

```python

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data
before = np.array([60, 55, 45, 50, 65, 70, 80, 85, 90, 40])
after = np.array([50, 55, 47, 50, 70, 73, 83, 79, 98, 45])

# Step 1: Calculate the differences (after - before)
differences = after - before

# Step 2: Calculate mean and standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(differences)  # Number of employees (pairs)

# Step 3: Calculate the t-statistic
t_statistic = mean_diff / (std_diff / np.sqrt(n))

# Step 4: Degrees of freedom (n - 1)
df = n - 1

# Step 5: Critical t-value for a one-tailed test at alpha=0.05
alpha = 0.05
critical_t = stats.t.ppf(1 - alpha, df)

# Step 6: Make the decision
if t_statistic > critical_t:
    decision = "Reject the null hypothesis: The training program improves typing speed."
else:
    decision = "Fail to reject the null hypothesis: No evidence that the training improves typing speed."

# Step 7: Plotting the differences
plt.figure(figsize=(8, 6))
plt.plot(range(1, n + 1), differences, marker='o', linestyle='-', color='b', label='Differences (After - Before)')
plt.axhline(0, color='red', linestyle='--', label='No Difference (0)')
plt.title('Differences in Typing Speed (After - Before) for 10 Employees')
plt.xlabel('Employee Index')
plt.ylabel('Typing Speed Difference')
plt.legend()
plt.grid(True)
plt.show()

# Output the results
print(f"Calculated t-statistic: {t_statistic:.2f}")
print(f"Critical t-value: {critical_t:.2f}")
print(f"Decision: {decision}")
```

```python
# chi-square test example
#it is used for goodness 
#  a fair die is rolled 120 times and folling results are obtained.
# face 1:22 times,20
# face 2:17 times,20
# face 3:20 times,20
# face 4:26 times,20
# face 5:22 times,20
# face 6:13 times,20
# test at alpha=5% whether is die is fair or not
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Observed frequencies
observed = np.array([22, 17, 20, 26, 22, 13])

# Expected frequencies (for a fair die, each face should appear 20 times)
expected = np.array([20, 20, 20, 20, 20, 20])

# Chi-square statistic calculation
chi_square_stat = np.sum((observed - expected) ** 2 / expected)

# Degree of freedom
df = len(observed) - 1

# Critical value from Chi-square distribution for alpha = 5% and df = 5
alpha = 0.05
chi_square_critical = chi2.ppf(1 - alpha, df)

# Decision
reject_null = chi_square_stat > chi_square_critical

# Results
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"Critical Value: {chi_square_critical:.4f}")
print(f"Reject Null Hypothesis: {reject_null}")

# Plotting the observed vs expected frequencies
faces = ['Face 1', 'Face 2', 'Face 3', 'Face 4', 'Face 5', 'Face 6']

plt.bar(faces, observed, alpha=0.7, label='Observed', color='blue')
plt.bar(faces, expected, alpha=0.7, label='Expected', color='orange')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies for Fair Die')
plt.legend()
plt.show()
```
